```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
# 取得する日の範囲を指定する
start = datetime.datetime(2020, 1, 1)
end = datetime.datetime(2020, 12, 31)

f = data.DataReader('AAPL', 'yahoo', start, end)

plt.title('AAPL from 2020.1.1 to 2020.12.31')

# fill_between でその日の最高値と最低値をプロットする
plt.fill_between(f.index, f['Low'], f['High'], color="b", alpha=0.5,)


# 始値をプロットする。
# 自動的に Index が Date になっているので、横軸が時間になる。


f['Open'].plot(figsize=(20,10))
plt.show()
```

脚注

アップルの株価の推移
アップルの株価と言えば、9月初頭に史上最高値を付けた後は続落していることが話題として挙がっています。
まず新製品発表イベント前後における同社株価の値動きについて見ていきます。
イベント発表前日である9/14における同社株価の終値は115.36ドルとなっていましたが、翌日15日の始値は118.74ドルとなっており、前日終値から3％ほど上昇しています。
これは主に新製品発表イベントに対する期待感から上昇したものと思われます。
イベントが開始した10時ごろの株価は118ドル前後でしたが、発表中から同社株価は続落し、発表後には116ドル台まで下落しました。
その後日中も軟調な推移が続き、15日におけるアップルの株価の終値は115.54ドルとなっていました。
また16日の執筆時時点では同社株価は一時113ドル台まで下落しており、軟調な推移が続いています。
アップルの株価は2020年初頭から80％以上上昇しており、ここ5年間で株価は4倍近くまで上昇しています。
非常に高いパフォーマンスを示しているものの、9月初頭からのナスダック総合指数の値動きからも分かる通り、調整局面に入っています。
アップルに関して述べるのであれば、新作iPhoneの発表が期待されていたにもかかわらず、発表されなかったことに対する失望は大きく、その事実が株価に対して反映されていると言えるでしょう。
またアップルの今後の成長も株価に織り込まれながら上昇していたものと言え、しばらくは軟調な推移が続く可能性があります。
今後のアップルの株価の推移を変えるものがあるとすれば、新しいiPhoneであるiPhone12の発表及びその売上でしょう。
アップルの株価は業績と連動しながら推移しているため、新製品の発表などに一喜一憂するのではなく、それらの製品がどれほどの利益を同社に対してもたらすのかどうかを常に意識する必要があると言えるのではないでしょうか。
